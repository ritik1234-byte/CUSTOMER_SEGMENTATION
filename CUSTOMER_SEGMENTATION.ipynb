```python
# K-Means Clustering
```

```python
data.columns
```

```python
x = data.iloc[:, [3,4]].values
x
```

```python
k = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i,init= 'k-means++', random_state = 0)
    kmeans.fit(x)
    k.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), k)
plt.show()
```

```python
model = KMeans(n_clusters = 6, init = 'k-means++', random_state = 0)
y_kemsn = model.fit_predict(x)
```

```python
plt.figure(1, figsize=(15, 8))

# Plot each cluster with labels
plt.scatter(x[y_kemsn == 0, 0], x[y_kemsn == 0, 1], s=100, c='yellow', label='Cluster 1: High Income, High Spend')
plt.scatter(x[y_kemsn == 1, 0], x[y_kemsn == 1, 1], s=100, c='blue', label='Cluster 2: High Income, Low Spend')
plt.scatter(x[y_kemsn == 2, 0], x[y_kemsn == 2, 1], s=100, c='pink', label='Cluster 3: Low Income,High Spend')
plt.scatter(x[y_kemsn == 3, 0], x[y_kemsn == 3, 1], s=100, c='green', label='Cluster 4: Low Income, Average Spend')
plt.scatter(x[y_kemsn == 4, 0], x[y_kemsn == 4, 1], s=100, c='red', label='Cluster 5: Low Income, Low Spend')
plt.scatter(x[y_kemsn == 5, 0], x[y_kemsn == 5, 1], s=100, c='orange', label='Cluster 6:Average income,High spend')

# Plot cluster centroids
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='black', label='Centroids')

# Add title and labels
plt.title('K Means Clustering Algorithm')
plt.xlabel('Annual Income (lakh)')
plt.ylabel('Spending Score (1-100)')

annotations = [
    (0, 'Low Income, Low Spend'),
    (1, 'Low Income, High Spend'),
    (2, 'High Income, Average Spend'),
    (3, 'Low Income, Average Spend'),
    (4, 'High Income, High Spend'),
    (5, 'High Income, Low Spend')
]

for i, text in annotations:
    plt.text(kmeans.cluster_centers_[i, 0], kmeans.cluster_centers_[i, 1], text,
             horizontalalignment='center', verticalalignment='center', fontsize=12, color='black',
             bbox=dict(facecolor='white', alpha=0.8, edgecolor='black'))

# Show plot
plt.show()
```
